# Logistic Regression: An Overview

Logistic Regression is a statistical method used for classification problems, where the goal is to predict the probability of an instance belonging to a particular category. Despite its name, it is a classification algorithm rather than a regression one.

## What is Logistic Regression?

### Definition
Logistic regression predicts the probability of an outcome belonging to a specific class using a sigmoid (logistic) function.

### Output
The model outputs probabilities (between 0 and 1), which can be mapped to binary (e.g., yes/no) or multiclass labels.

### Key Equation
\[
P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + ... + \beta_nX_n)}}
\]

- **\(P(Y=1|X)\):** Probability of the positive class.
- **\(\beta_0, \beta_1, ..., \beta_n\):** Coefficients learned from the data.
- **\(X_1, X_2, ..., X_n\):** Input features.

## Algorithm Basics

### 1. Data Preprocessing
- Handle missing values.
- Scale numerical features.
- Encode categorical variables.

### 2. Initialize Parameters
- Start with random weights (\(\beta\)).

### 3. Forward Propagation
- Compute the weighted sum of inputs: \(Z = \beta_0 + \beta_1X_1 + ...\).
- Apply the sigmoid function to get probabilities.

### 4. Loss Function
Use the log-loss (binary cross-entropy):
\[
L = -\frac{1}{m} \sum_{i=1}^m \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right]
\]

### 5. Gradient Descent
- Update weights iteratively to minimize the loss.

### 6. Prediction
- Classify based on a threshold (e.g., 0.5 for binary classification).

## Use Cases

### Binary Classification
- **Spam detection:** Spam/Not Spam
- **Medical diagnosis:** Disease/No Disease

### Multiclass Classification
With extensions like One-vs-All or Softmax:
- Handwriting digit recognition.

### Predicting Probabilities
- **Risk assessment:** e.g., loan default probability.

### Log-odds Interpretation
Logistic regression coefficients can explain feature importance.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("sample.csv")
```

```python
data.head()
```

```python
plt.scatter(data.age,data.job)
plt.xlabel('age')
```

```python
x = data[['age']]
```

```python
y = data[['job']]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.1)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python

# get predictions from model
model.predict(x_test)
```

```python

# accuracy of model
model.score(x_test,y_test)
```

```python

age = np.array([[20],[25],[30],[35]])
model.predict(age)
```
